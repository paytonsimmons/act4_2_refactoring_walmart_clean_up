```python
TTPQ_Pre_FV = pd.read_csv('data/TPQ_Pre.csv')
TTPQ_Post_FV = pd.read_csv('data/TPQ_Post.csv')
```

```python
TTPQ_Pre_FV2 = (TTPQ_Pre_FV >> select(X['indicator'], X['Q1'], X['Q2'], X['Q3'], X['Q4'], X['Q5'], X['Q6'], X['Q7'], X['Q8'], X['Q9'],
                                          X['Q10'], X['Q11'], X['Q12'], X['Q13'], X['Q14'], X['Q15'], X['Q16'], X['Q17'], X['Q18'], X['Q19'],
                                          X['Q20'], X['Q21'], X['Q22'], X['Q23'], X['Q24'], X['Q25'], X['Q26'], X['Q27'], X['Q28'], X['Q29'],
                                          X['Q30'], X['Q31'], X['Q32'], X['Q33'], X['Q34'], X['Q35']))
```

```python
TTPQ_Post_FV2 = (TTPQ_Post_FV >> select(X['indicator'], X['Q1'], X['Q2'], X['Q3'], X['Q4'], X['Q5'], X['Q6'], X['Q7'], X['Q8'], X['Q9'],
                                          X['Q10'], X['Q11'], X['Q12'], X['Q13'], X['Q14'], X['Q15'], X['Q16'], X['Q17'], X['Q18'], X['Q19'],
                                          X['Q20'], X['Q21'], X['Q22'], X['Q23'], X['Q24'], X['Q25'], X['Q26'], X['Q27'], X['Q28'], X['Q29'],
                                          X['Q30'], X['Q31'], X['Q32'], X['Q33'], X['Q34'], X['Q35']))
```

```python
(TTPQ_Pre_FV2
>> mutate(Pre_Post = "Pre"))
```

```python
(TTPQ_Post_FV2
>> mutate(Pre_Post = "Post"))
```

```python
TTPQ_Joined = pd.concat([TTPQ_Pre_FV2,TTPQ_Post_FV2]).sort_values('indicator')
TTPQ_Final_Table = TTPQ_Joined.melt(id_vars=["indicator", "Pre_Post"])
TTPQ_Final_Table
```

```python
new_scale = {'Strongly Disagree':1,'Disagree':2,'Neutral':3,'Agree':4,'Strongly Agree':5}

TTPQ = (TTPQ_Final_Table
>> mutate(Answer = X.value.map(new_scale))
>> mutate(Question = X.variable)
>> mutate(Pre_Post = X.Pre_Post.apply(lambda x: 'Pre' if x == 'Pre' else 'Post'))
>> select(X.Pre_Post, X.indicator, X.Question, X.Answer)
>> head())
```

```python
TTPQ.to_csv('TTPQ.csv')
```
